```julia
# Joseph Loy EE104 HW1
using Plots

# Question 2a
N_samp = 30;
N_points = 500;
X = rand(N_samp);
Y = sin.(10 * X);
x = range(0, stop=1, length=N_points);
f_x = sin.(10*x);
scatter(X, Y, label ="Sin(10x) , 30 rand points")
plot!(x, f_x, label ="Sin(10x)")

```

```julia
function knn(X, Y, x, k)
    n = size(X)[1]
    dists_sq = [sum((X[i,:] .- x).^2) for i=1:n] #find distances of examples to x
    nearest_neighbor_idxs = sortperm(dists_sq)[1:k] #find k-nearest neighbors
    y_hat = sum(Y[nearest_neighbor_idxs, :], dims=1)/k #average k-nearest neighbors
    return y_hat
end
```

```julia
function softknn(X, Y, x, rho)
    n = size(X)[1]
    exp_weights = [exp(-sum((X[i,:] .- x).^2)/rho^2) for i=1:n]
    w = exp_weights / sum(exp_weights) #find weights
    y_hat = sum(w .* Y, dims=1) #weighted combination of Y
    return y_hat
end
```

```julia
# Question 2b k = 1
N_samp = 30
N_points = 500
X = rand(N_samp)
Y = sin.(10 * X)
x = range(0, stop=1, length=N_points)
f_x = sin.(10*x)

test = .5
r1 = knn(X,Y, test, 1)
scatter(X,Y)
scatter!([test], [r1])


```
