```julia
# EE104 HW6
using JSON
using LinearAlgebra
using Plots
using Flux
using Random

include("readclassjson.jl")
include("regression_fit.jl")
include("nn_regression.jl")

data1 = readclassjson("non_quad_regs.json")
data2 = readclassjson("nn_regression.json")

U_tr1 = data1["U_train"]
v_tr1 = data1["v_train"]
U_te1 = data1["U_test"]
v_te1 = data1["v_test"]

U_tr2 = data2["U_train"]
v_tr2 = data2["v_train"]
U_te2 = data2["U_test"]
v_te2 = data2["v_test"]
```

```julia
# Create regularization functions for question 1
function con_r(theta)
    return 0
end

function quad_r(theta)
    return sum(theta.^2)
end

function abs_r(theta)
    return sum(abs.(theta))
end

function squ_r(theta)
    return sqrt(sum(abs.(theta)))
end

function quad_l(y_hat, y)
    return sum((y_hat - y).^2)
end
```

```julia
# Train regression models
lam = 10 .^ range(-3, stop=1, length=20)

thetas_quadR = zeros(length(lam), 30)
thetas_absR = zeros(length(lam), 30)
thetas_squR = zeros(length(lam), 30)




for i = 1:length(lam)
    thetas_quadR[i, :] = regression_fit(U_tr1, v_tr1, quad_l, quad_r, lam[i])
    thetas_absR[i, :] = regression_fit(U_tr1, v_tr1, quad_l, abs_r, lam[i])
    thetas_squR[i, :] = regression_fit(U_tr1, v_tr1, quad_l, squ_r, lam[i])
end
theta_con = regression_fit(U_tr1, v_tr1, quad_l, con_r, lam[1])
```

```julia
# Calculate RMS errors for predictors for training data
RMSE_quad = zeros(length(lam))
RMSE_abs = zeros(length(lam))
RMSE_squ = zeros(length(lam))
RMSE_quad_test = zeros(length(lam))
RMSE_abs_test = zeros(length(lam))
RMSE_squ_test = zeros(length(lam))
RMSE_con = norm((U_tr1 * theta_con) .- v_tr1) / sqrt(length(v_tr1))
RMSE_con_test = norm((U_te1 * theta_con) .- v_te1) / sqrt(length(v_te1))

for i = 1:length(lam)
    RMSE_quad[i] = norm((U_tr1 * thetas_quadR[i, :]) .- v_tr1) / sqrt(length(v_tr1))
    RMSE_abs[i] = norm((U_tr1 * thetas_absR[i, :]) .- v_tr1) / sqrt(length(v_tr1))
    RMSE_squ[i] = norm((U_tr1 * thetas_squR[i, :]) .- v_tr1) / sqrt(length(v_tr1))
    
    RMSE_quad_test[i] = norm((U_te1 * thetas_quadR[i, :]) .- v_te1) / sqrt(length(v_te1))
    RMSE_abs_test[i] = norm((U_te1 * thetas_quadR[i, :]) .- v_te1) / sqrt(length(v_te1))
    RMSE_squ_test[i] = norm((U_te1 * thetas_quadR[i, :]) .- v_te1) / sqrt(length(v_te1))
end
```

```julia
# Plot training errors for each of the lambdas
plot(lam, [RMSE_quad, RMSE_quad_test])
xlabel!("lambda")
ylabel!("RMSE")
title!("RMSE quadratic regulariztion")
```

```julia
# Plot training errors for each of the lambdas
plot(lam, [RMSE_abs, RMSE_abs_test])
xlabel!("lambda")
ylabel!("RMSE")
title!("RMSE absolute regulariztion")
```

```julia
# Plot training errors for each of the lambdas
plot(lam, [RMSE_squ, RMSE_squ_test])
xlabel!("lambda")
ylabel!("RMSE")
title!("RMSE absolute root regulariztion")
```

```julia
# Plot training errors for each of the lambdas
plot(lam, [fill(RMSE_con, length(lam)), fill(RMSE_con_test, length(lam))])
xlabel!("lambda")
ylabel!("RMSE")
title!("RMSE quadratic regulariztion")
```

The model that performs the best is the absolute regularization model. This likely means that there are some outliers in the data that cause the quadratic model to perform poorly. The absolute regularization and absolute root regularization perform approximately the same.
